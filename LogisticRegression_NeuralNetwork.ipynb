**adding win/loss column**

```python
condlist = [
    final8_df['Team1Score'] > final8_df['Team2Score'],
    final8_df['Team1Score'] < final8_df['Team2Score']
]

win_loss = [1,0]

final8_df['Team1Win'] = np.select(condlist, win_loss, 1)

final8_df.head()
```

**clean new dataframe**

```python
finals_cleaned_df = finals_df.drop(['Team1', 'Team1Score', 'Team2', 'Team2Score', 'FULL NAME', 'AGE', 'POS'], axis = 1)

finals_cleaned_df.sample(10)
```

```python
finals_cleaned_df = finals_cleaned_df.rename(columns = {'Team1Win':'TeamWin'})

finals_cleaned_df.sample(5)
```

```python
finals_cat = finals_cleaned_df.dtypes[finals_cleaned_df.dtypes == 'object'].index.tolist()
finals_cat
```

```python
finals_cleaned_df[finals_cat].nunique()
```

```python
enc = OneHotEncoder(sparse=False)

encode_df = pd.DataFrame(enc.fit_transform(finals_cleaned_df[finals_cat]))

encode_df.columns = enc.get_feature_names(finals_cat)

encode_df.sample(10)
```

```python
finals_cleaned_df = finals_cleaned_df.merge(encode_df, left_index=True, right_index=True)

finals_cleaned_df = finals_cleaned_df.drop(finals_cat, 1)

finals_cleaned_df.sample(10)
```

```python
finals_cleaned_df.describe()
```

## Separate Features From the Target

```python
y = finals_cleaned_df.TeamWin.values
X = finals_cleaned_df.drop(columns = ['TeamWin', 'GP', 'FTA', '2PA']).values
```

```python
# lowered accuracy scores
# y = finals_cleaned_df.TeamWin.values
# X = finals_cleaned_df.drop(
#   columns = ['TeamWin', 'GP', 'TEAM_Boston Celtics', 'TEAM_Dallas Mavericks',
#           'TEAM_Golden State Warriors', 'TEAM_Memphis Grizzlies',
#              'TEAM_Miami Heat', 'TEAM_Milwaukee Bucks',
#              'TEAM_Philadelphia 76ers', 'TEAM_Phoenix Suns']).values
```

## Split Data Into Training and Testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2)
```

## Standardize the Features

```python
scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Logistic Regression Model

```python
# instantiate logistic regression model
model = LogisticRegression(max_iter=200, random_state = 42)

# fit and train the model
model.fit(X_train_scaled, y_train)

# make predictions
pred = model.predict(X_test_scaled)
```

```python
# accuracy score
print(f'Logistic Regression Model Accuracy Score: {accuracy_score(y_test, pred)}')
```

```python
# confusion matrix
print('Confusion Matrix:')
print(confusion_matrix(y_test, pred))
```

```python
# classification report
print('Classification Report:')

print(classification_report(y_test, pred, target_names = ['lose', 'win']))
```

## Neural Network

```python
# Define the model - deep neural net
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 20   #75 previous test
hidden_nodes_layer2 = 10   #35 previous test
hidden_nodes_layer3 =  5   #10 previous test

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units = hidden_nodes_layer1, activation = 'relu' , input_dim = number_input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units = hidden_nodes_layer2, activation = 'relu'))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units = hidden_nodes_layer3, activation = 'relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs = 100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
print(f'Loss: {model_loss}, Accuracy: {model_accuracy}')
```
